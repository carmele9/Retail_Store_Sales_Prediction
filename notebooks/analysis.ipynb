```python
%matplotlib inline 
```

```python
df_category = df.groupby('Category')['Amount'].sum().sort_values(ascending=False).reset_index()

sns.barplot(data = df_category, x = 'Category', y = 'Amount', hue='Category', palette='coolwarm', legend=False)
plt.xticks(rotation = 45)
plt.xlabel('Category')
plt.ylabel('Amount')
plt.title('Cuales son las categorias más compradas?')
plt.tight_layout()
plt.show()
```

```python
df_gender = df.groupby('Gender')['Amount'].sum().sort_values(ascending=False).reset_index()

sns.barplot(data = df_gender, x = 'Gender', y = 'Amount', hue='Gender', palette='coolwarm', legend=False)
plt.xticks(rotation = 45)
plt.xlabel('Gender')
plt.ylabel('Amount')
plt.title('Qué género compra más?')
plt.tight_layout()
plt.show()
```

```python
df_payment = df.groupby('PaymentMethod')['Amount'].sum().sort_values(ascending=False).reset_index()

sns.barplot(data = df_payment, x = 'PaymentMethod', y = 'Amount', hue='PaymentMethod', palette='coolwarm', legend=False)
plt.xticks(rotation = 45)
plt.xlabel('PaymentMethod')
plt.ylabel('Amount')
plt.title('Cual es el método más usado para comprar?')
plt.tight_layout()
plt.show()
```

```python
df_season = df.groupby('Season')['Amount'].sum().sort_values(ascending=False).reset_index()

sns.barplot(data = df_season, x = 'Season', y = 'Amount', hue='Season', palette='coolwarm', legend=False)
plt.xticks(rotation = 45)
plt.xlabel('Season')
plt.ylabel('Amount')
plt.title('En cual estacion del Año se compra más?')
plt.tight_layout()
plt.show()
```

```python
numerical = df.select_dtypes(include='number').columns.to_list()
numerical.pop(0)

sns.boxplot(df[numerical], palette='coolwarm', legend=False)
plt.xticks(rotation = 45)
plt.show()
```

```python
corr = df[numerical].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Correlation Matrix")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```
